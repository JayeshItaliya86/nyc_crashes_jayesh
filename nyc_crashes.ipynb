```python
# Import libraries
import pandas as pd
import numpy as np
pd.set_option("Display.max_columns", None)
pd.set_option('max_colwidth', 5000)
```

```python
# To read the csv file
nyc_motor_vehical_crashes = pd.read_csv('./data_100000.csv',
                                        parse_dates=[['crash_date', 'crash_time']],
                                        skipinitialspace=True)
df = nyc_motor_vehical_crashes.copy()
```

```python
# create the dictionary with the old and new names
new_names = {'crash_date_crash_time': 'crash_date_time',
             'vehicle_type_code1': 'vehicle_type_code_1',
             'vehicle_type_code2': 'vehicle_type_code_2'}

# Rename the columns
df = df.rename(columns=new_names)
```

```python
# To change empty string to np.nan values
df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# Change data type from float64 to int64
df['zip_code'] = (df['zip_code'].fillna(0)
                  .astype('Int64')
                  .where(df['zip_code'].notnull()))
```

```python
# Drop some columns
df = df.drop(columns=['borough', 'location', 'on_street_name',
                      'cross_street_name', 'off_street_name', 
                      'collision_id'], axis=1)
```

```python
# To drop the duplicate rows
df = df.drop_duplicates()
```

```python
# To narrow some factor of contributing factor 
def narrowing_down_factor(factor):
    distraction = ['Driver Inattention/Distraction','Passenger Distraction','Fell Asleep','Outside Car Distraction',
                   'Fatigued/Drowsy','Cell Phone (hand-Held)','Using On Board Navigation Device',
                   'Tinted Windows','Eating or Drinking', 'Other Electronic Device', 'Cell Phone (hands-free)',
                   'Listening/Using Headphones']
    driver_mistake = ['Following Too Closely','Failure to Yield Right-of-Way','Backing Unsafely',
                      'Passing or Lane Usage Improper','Passing Too Closely','Unsafe Lane Changing',
                      'Turning Improperly','Driver Inexperience', 'Failure to Keep Right',
                      'Driverless/Runaway Vehicle', 'Oversized Vehicle']
    illegal_action = ['Unsafe Speed','Alcohol Involvement','Traffic Control Disregarded','Aggressive Driving/Road Rage',
                      'Drugs (illegal)']
    other_involvement =  ['Other Vehicular','Reaction to Uninvolved Vehicle', 'Animals Action',
                          'View Obstructed/Limited','Pedestrian/Bicyclist/Other Pedestrian Error/Confusion',
                          'Glare','Vehicle Vandalism']
    bad_road_infrastructure = ['Pavement Slippery', 'Obstruction/Debris','Pavement Defective',
                               'Other Lighting Defects']
    car_failures = ['Brakes Defective', 'Steering Failure', 'Tire Failure/Inadequate',
                    'Traffic Control Device Improper/Non-Working', 'Lane Marking Improper/Inadequate',
                    'Tow Hitch Defective','Headlights Defective', 'Shoulders Defective/Improper',
                    'Windshield Inadequate']
    medical = ['Lost Consciousness', 'Illnes','Accelerator Defective','Physical Disability','Prescription Medication']
    unspecified = ['Unspecified']
    if factor in unspecified:
        return "unspecified"
    elif factor in medical:
        return "medical"
    elif factor in car_failures:
        return "car_failure"
    elif factor in bad_road_infrastructure:
        return "bad_road_infrastructure"
    elif factor in other_involvement:
        return "other_involvement"
    elif factor in illegal_action:
        return "illegal_action"
    elif factor in driver_mistake:
        return "driver_mistake"
    elif factor in distraction:
        return "distraction"
```

```python
# To apply fuction on dataset to get combine factors
df['contributing_factor_1'] = df['contributing_factor_vehicle_1'].apply(lambda x: narrowing_down_factor(x))
df['contributing_factor_2'] = df['contributing_factor_vehicle_2'].apply(lambda x: narrowing_down_factor(x))
df['contributing_factor_3'] = df['contributing_factor_vehicle_3'].apply(lambda x: narrowing_down_factor(x))
df['contributing_factor_4'] = df['contributing_factor_vehicle_4'].apply(lambda x: narrowing_down_factor(x))
df['contributing_factor_5'] = df['contributing_factor_vehicle_5'].apply(lambda x: narrowing_down_factor(x))
```

```python
df.head()
```

```python
# To apply one hot encoding
contribute_factor_1 = pd.get_dummies(df.contributing_factor_1, prefix='contributing_factor_1', drop_first="True")
contribute_factor_2 = pd.get_dummies(df.contributing_factor_2, prefix='contributing_factor_2', drop_first="True")
contribute_factor_3 = pd.get_dummies(df.contributing_factor_3, prefix='contributing_factor_3', drop_first="True")
contribute_factor_4 = pd.get_dummies(df.contributing_factor_4, prefix='contributing_factor_4', drop_first="True")
contribute_factor_5 = pd.get_dummies(df.contributing_factor_5, prefix='contributing_factor_5', drop_first="True")
```

```python
df1 = pd.concat([contribute_factor_1], axis=1)
df2 = pd.concat([contribute_factor_2], axis=1)
df3 = pd.concat([contribute_factor_3], axis=1)
df4 = pd.concat([contribute_factor_4], axis=1)
df5 = pd.concat([contribute_factor_5], axis=1)
```

```python
result = pd.concat([df, df1, df2, df3, df4, df5], axis=1)
```
